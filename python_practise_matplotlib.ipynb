```python
import matplotlib.pyplot as plt
import numpy as np
import math
data = [-1, -4.3, 15, 21, 31]
plt.plot(data)
plt.show()
plt.plot(data, "o--b")
plt.show()
```

```python
days = range(1, 7)
celsius = [25.6, 23.2, 18.5, 28.3, 26.5, 30.5]
plt.plot(days, celsius)
plt.show()
```

```python
x = np.linspace(0, 2*math.pi, 1000)
sinus = np.sin(x)
cosinus = np.cos(x)
plt.plot(x,sinus,label="sin")
plt.plot(x, cosinus,label="cos")
plt.show()
plt.plot(sinus, cosinus)
plt.show()
print(sinus.dot(cosinus))
```

```python
days = range(1, 7)
celsius = [25.6, 23.2, 18.55, 28.3, 25.7, 30.5]
plt.plot(days, celsius, "r--o")
plt.show()
plt.plot(days, celsius, "^")
plt.xlabel("days")
plt.ylabel("Temperature")
plt.grid(True)
plt.show()
```
